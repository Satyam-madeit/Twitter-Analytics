# Data Exploration

This notebook is the starting point for exploring the tweet datasets in the `Datasets/` folder before building and comparing machine learning models in the other notebooks.

The goal here is to understand the structure, quality, and useful signals in the available data. This exploration should help guide later decisions about feature engineering, preprocessing, target selection, and model evaluation.

## Planned exploration

- Load and inspect the available JSON datasets: `london.json`, `new_york.json`, `paris.json`, and `random_tweets.json`.
- Review the fields available for each tweet and check whether the schema is consistent across datasets.
- Identify missing values, duplicated records, unusual values, and potential data quality issues.
- Explore text-related features such as tweet length, hashtags, mentions, links, and language patterns.
- Compare city-based datasets with the random tweet dataset to understand distribution differences.
- Summarize observations that may be useful for downstream ML notebooks.

## Notes for ML work

Findings from this notebook should be used to decide which columns are reliable, which transformations are needed, and which prediction tasks are realistic for the data. Any assumptions or discovered limitations should be documented here before they are used in model-focused notebooks.
